```python
import discord
import re
from fuzzywuzzy import fuzz
import nest_asyncio
import asyncio
import os
from dotenv import load_dotenv
import random
import threading

# Chargement des variables d'environnement
load_dotenv()
token = os.getenv("DISCORD_TOKEN")
if token is None:
    raise ValueError("âš ï¸ La variable d'environnement DISCORD_TOKEN n'est pas dÃ©finie.")
print("Token chargÃ© avec succÃ¨s âœ…")

# Correction de la boucle Jupyter si besoin
nest_asyncio.apply()

# Intents Discord
intents_discord = discord.Intents.default()
intents_discord.messages = True
client = discord.Client(intents=intents_discord)

# Dictionnaire dâ€™intents (chansons)
intents = {
    "chanter": {
        "Sardou": {
            "titre": "Les Lacs du Connemara",
            "paroles": "Câ€™est en Irlande quâ€™il est nÃ©, entre Galway et Limerick..."
        },
        "Renaud": {
            "titre": "Mistral gagnant",
            "paroles": "Ã€ m'asseoir sur un banc cinq minutes avec toi..."
        },
        "Steve": {
            "titre": "Le Poasson Steve",
            "paroles": "Steve, le poisson... ORANGeuuuh !!!"
        },
        "Dalida": {
            "titre": "Gigi l'Amoroso",
            "paroles": "Gigi l'Amoroso, croqueur d'amour..."
        },
        "Johnny": {
            "titre": "Que je t'aime",
            "paroles": "Quand tes cheveux s'Ã©talent..."
        },
        "Brel": {
            "titre": "Ne me quitte pas",
            "paroles": "Ne me quitte pas, il faut oublier..."
        },
        "Aznavour": {
            "titre": "La BohÃ¨me",
            "paroles": "La BohÃ¨me, la BohÃ¨me..."
        },
        "Claude FranÃ§ois": {
            "titre": "Comme d'habitude",
            "paroles": "Je me lÃ¨ve et je te bouscule..."
        },
        "Jean-Jacques Goldman": {
            "titre": "Encore un matin",
            "paroles": "Encore un matin, un matin pour rien..."
        },
        "Daniel Balavoine": {
            "titre": "L'Aziza",
            "paroles": "Ton nom, il est doux Ã  murmurer..."
        },
        "MylÃ¨ne Farmer": {
            "titre": "DÃ©senchantÃ©e",
            "paroles": "Tout est chaos, Ã  cÃ´tÃ©..."
        },
        "Michel Berger": {
            "titre": "La groupie du pianiste",
            "paroles": "Elle passe ses nuits sans dormir..."
        }
    }
}

# Fonction de reconnaissance d'intention
def get_intent(user_input, intents, threshold=80):
    user_input = user_input.lower()

    if "chante" in user_input or "chanson" in user_input or "steve" in user_input or "paroles" in user_input:
        for artiste in intents["chanter"]:
            if artiste.lower() in user_input:
                return "chanter", artiste
        return "chanter", None

    best_score = 0
    best_intent = "intent_inconnu"
    for intent, keywords in intents.items():
        if isinstance(keywords, dict):
            continue
        for keyword in keywords:
            score = fuzz.partial_ratio(user_input, keyword.lower())
            if score > best_score and score >= threshold:
                best_score = score
                best_intent = intent

    return best_intent, None

# Quand le bot est prÃªt
@client.event
async def on_ready():
    print("âœ… Le bot est prÃªt et connectÃ© !")

# Quand un message est reÃ§u
@client.event
async def on_message(message):
    if message.author == client.user:
        return

    mess = message.content.lower().strip()
    print(f"ğŸ“¨ Message reÃ§u : {mess}")

    if mess == "steve":
        await message.channel.send("Oui câ€™est moi, veux-tu que je chante ? ğŸ¶")
        return

    if mess.startswith('steve'):
        pattern = r'^steve\s*'
        cleaned_message = re.sub(pattern, '', mess, flags=re.IGNORECASE)
        detected_intent, artiste = get_intent(cleaned_message, intents)
        print(f"ğŸ¯ Intent : {detected_intent}, Artiste : {artiste}")

        if detected_intent == "chanter":
            if artiste and artiste in intents["chanter"]:
                chanson = intents["chanter"][artiste]
                await message.channel.send(
                    f"ğŸµ **{artiste} â€“ {chanson['titre']}**\n\n{chanson['paroles']}"
                )
            else:
                await message.channel.send("Je ne connais pas encore cet artiste ğŸ˜¥.")
        else:
            await message.channel.send(f"Intent reconnu : **{detected_intent}**")

    else:
        detected_intent, artiste = get_intent(mess, intents)
        print(f"ğŸ•µï¸â€â™‚ï¸ (Sans steve) Intent : {detected_intent}, Artiste : {artiste}")

        if detected_intent == "chanter":
            if artiste and artiste in intents["chanter"]:
                chanson = intents["chanter"][artiste]
                await message.channel.send(
                    f"ğŸ¤ Je suis Steve, et j'ai devinÃ© que tu voulais :\n"
                    f"ğŸµ **{artiste} â€“ {chanson['titre']}**\n\n{chanson['paroles']}"
                )
            else:
                await message.channel.send("Tu veux que je chante ? Je suis prÃªt, mais je ne connais pas cet artiste ğŸ˜¥.")
        else:
            artiste_aleatoire = random.choice(list(intents["chanter"].keys()))
            chanson = intents["chanter"][artiste_aleatoire]
            await message.channel.send(
                f"ğŸ‘€ On ne mâ€™a pas appelÃ© mais je vais chanter quand mÃªme, je suis **Steve** ! ğŸ¤\n"
                f"ğŸµ **{artiste_aleatoire} â€“ {chanson['titre']}**\n\n{chanson['paroles']}"
            )

# Fonction pour exÃ©cuter dans un Jupyter ou script
def run_bot():
    client.run(token)

# Lance dans un thread (si Jupyter)
thread = threading.Thread(target=run_bot)
thread.start()
```
