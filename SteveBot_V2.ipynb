```python
import discord
import re
from fuzzywuzzy import fuzz
import nest_asyncio
import asyncio
import os
from dotenv import load_dotenv
import random
import threading

# Chargement des variables d'environnement
load_dotenv()
token = os.getenv("DISCORD_TOKEN")
if token is None:
    raise ValueError("⚠️ La variable d'environnement DISCORD_TOKEN n'est pas définie.")
print("Token chargé avec succès ✅")

# Correction de la boucle Jupyter si besoin
nest_asyncio.apply()

# Intents Discord
intents_discord = discord.Intents.default()
intents_discord.messages = True
client = discord.Client(intents=intents_discord)

# Dictionnaire d’intents (chansons)
intents = {
    "chanter": {
        "Sardou": {
            "titre": "Les Lacs du Connemara",
            "paroles": "C’est en Irlande qu’il est né, entre Galway et Limerick..."
        },
        "Renaud": {
            "titre": "Mistral gagnant",
            "paroles": "À m'asseoir sur un banc cinq minutes avec toi..."
        },
        "Steve": {
            "titre": "Le Poasson Steve",
            "paroles": "Steve, le poisson... ORANGeuuuh !!!"
        },
        "Dalida": {
            "titre": "Gigi l'Amoroso",
            "paroles": "Gigi l'Amoroso, croqueur d'amour..."
        },
        "Johnny": {
            "titre": "Que je t'aime",
            "paroles": "Quand tes cheveux s'étalent..."
        },
        "Brel": {
            "titre": "Ne me quitte pas",
            "paroles": "Ne me quitte pas, il faut oublier..."
        },
        "Aznavour": {
            "titre": "La Bohème",
            "paroles": "La Bohème, la Bohème..."
        },
        "Claude François": {
            "titre": "Comme d'habitude",
            "paroles": "Je me lève et je te bouscule..."
        },
        "Jean-Jacques Goldman": {
            "titre": "Encore un matin",
            "paroles": "Encore un matin, un matin pour rien..."
        },
        "Daniel Balavoine": {
            "titre": "L'Aziza",
            "paroles": "Ton nom, il est doux à murmurer..."
        },
        "Mylène Farmer": {
            "titre": "Désenchantée",
            "paroles": "Tout est chaos, à côté..."
        },
        "Michel Berger": {
            "titre": "La groupie du pianiste",
            "paroles": "Elle passe ses nuits sans dormir..."
        }
    }
}

# Fonction de reconnaissance d'intention
def get_intent(user_input, intents, threshold=80):
    user_input = user_input.lower()

    if "chante" in user_input or "chanson" in user_input or "steve" in user_input or "paroles" in user_input:
        for artiste in intents["chanter"]:
            if artiste.lower() in user_input:
                return "chanter", artiste
        return "chanter", None

    best_score = 0
    best_intent = "intent_inconnu"
    for intent, keywords in intents.items():
        if isinstance(keywords, dict):
            continue
        for keyword in keywords:
            score = fuzz.partial_ratio(user_input, keyword.lower())
            if score > best_score and score >= threshold:
                best_score = score
                best_intent = intent

    return best_intent, None

# Quand le bot est prêt
@client.event
async def on_ready():
    print("✅ Le bot est prêt et connecté !")

# Quand un message est reçu
@client.event
async def on_message(message):
    if message.author == client.user:
        return

    mess = message.content.lower().strip()
    print(f"📨 Message reçu : {mess}")

    if mess == "steve":
        await message.channel.send("Oui c’est moi, veux-tu que je chante ? 🎶")
        return

    if mess.startswith('steve'):
        pattern = r'^steve\s*'
        cleaned_message = re.sub(pattern, '', mess, flags=re.IGNORECASE)
        detected_intent, artiste = get_intent(cleaned_message, intents)
        print(f"🎯 Intent : {detected_intent}, Artiste : {artiste}")

        if detected_intent == "chanter":
            if artiste and artiste in intents["chanter"]:
                chanson = intents["chanter"][artiste]
                await message.channel.send(
                    f"🎵 **{artiste} – {chanson['titre']}**\n\n{chanson['paroles']}"
                )
            else:
                await message.channel.send("Je ne connais pas encore cet artiste 😥.")
        else:
            await message.channel.send(f"Intent reconnu : **{detected_intent}**")

    else:
        detected_intent, artiste = get_intent(mess, intents)
        print(f"🕵️‍♂️ (Sans steve) Intent : {detected_intent}, Artiste : {artiste}")

        if detected_intent == "chanter":
            if artiste and artiste in intents["chanter"]:
                chanson = intents["chanter"][artiste]
                await message.channel.send(
                    f"🎤 Je suis Steve, et j'ai deviné que tu voulais :\n"
                    f"🎵 **{artiste} – {chanson['titre']}**\n\n{chanson['paroles']}"
                )
            else:
                await message.channel.send("Tu veux que je chante ? Je suis prêt, mais je ne connais pas cet artiste 😥.")
        else:
            artiste_aleatoire = random.choice(list(intents["chanter"].keys()))
            chanson = intents["chanter"][artiste_aleatoire]
            await message.channel.send(
                f"👀 On ne m’a pas appelé mais je vais chanter quand même, je suis **Steve** ! 🎤\n"
                f"🎵 **{artiste_aleatoire} – {chanson['titre']}**\n\n{chanson['paroles']}"
            )

# Fonction pour exécuter dans un Jupyter ou script
def run_bot():
    client.run(token)

# Lance dans un thread (si Jupyter)
thread = threading.Thread(target=run_bot)
thread.start()
```
